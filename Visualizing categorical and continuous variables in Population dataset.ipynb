## <span style="color:#9932CC"><strong>Adding relevant columns to the dataset</strong></span>

#### We can calculate **percentage increase in population for each decade** like 1960-1970, 1970-1980, ... etc upto 2020-2023.

```python
years = ['1960', '1970', '1980', '1990', '2000', '2010', '2020', '2023']
df[years] = df[years].apply(pd.to_numeric, errors='coerce')
```

```python
# Calculate percentage increases for each decade
for i in range(len(years) - 1):
    start_year = years[i]
    end_year = years[i + 1]
    column_name = f"{start_year}-{end_year} Increase (%)"
    df[column_name] = ((df[end_year] - df[start_year]) / df[start_year]) * 100

# Display the updated DataFrame
print(df.head())
```

## <span style="color:#9932CC"><strong>Separate categorical and continuous variables</strong></span>

```python
categorical_vars = df.select_dtypes(include=['object', 'category']).columns.tolist()
continuous_vars = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
```

```python
# Display the results
print("Categorical Variables:", categorical_vars)
print("Continuous Variables:", continuous_vars)
```

## <span style="color:#9932CC"><strong>Create a bar chart to find the top 10 countries having highest population in the year 2023.</strong></span>

```python
latest_year = '2023'

# Sort by population and select the top 10 countries
top_countries = df.sort_values(by=latest_year, ascending=False).head(10)
```

```python
# Extract data for the bar chart
countries = top_countries['Country Name']
populations = top_countries[latest_year]

# Plot the bar chart
plt.figure(figsize=(12, 8))
plt.bar(countries, populations, color='blue')
plt.title("Top 10 Countries by Population (2023)", fontsize=16)
plt.xlabel("Country", fontsize=14)
plt.ylabel("Population", fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12) 
plt.tight_layout()  
plt.show()
```

#### 
* **IDA & IBRD** total reports highest population in 2023.
* **Low and middle income** countries reports the second highest population in 2023.

## <span style="color:#9932CC"><strong>List the top 10 countries based on their population increase in each decade</strong></span>

```python
# List of decade columns with population increase percentages
decade_columns = ['1960-1970 Increase (%)', '1970-1980 Increase (%)', '1980-1990 Increase (%)', 
                  '1990-2000 Increase (%)', '2000-2010 Increase (%)', '2010-2020 Increase (%)', '2020-2023 Increase (%)']

# Dictionary to store top 10 countries for each decade
top_10_countries = {}

# Loop through each decade column
for decade in decade_columns:
    # Sort the countries based on the percentage increase and get the top 10
    top_10_countries[decade] = df.nlargest(10, decade)[['Country Name', decade]]

# Display the top 10 countries for each decade
for decade, data in top_10_countries.items():
    print(f"Top 10 countries for {decade}:")
    print(data)
    print("\n")
```

## <span style="color:#9932CC"><strong>Create multiple bar charts representing top 10 countries based on their population increase in each decade</strong></span>

```python
# Create a figure with multiple subplots (one for each decade)
plt.figure(figsize=(15, 10))

# Loop through each decade column to plot the bar charts for top 10 countries
for i, decade in enumerate(decade_columns, 1):
    # Get the top 10 countries for this decade
    top_10 = top_10_countries[decade]
    
    # Plot bar chart for the population increase of top 10 countries in this decade
    plt.subplot(3, 3, i)  # Create a subplot for each decade (3 rows, 3 columns)
    plt.bar(top_10['Country Name'], top_10[decade], color='skyblue', edgecolor='black')
    plt.title(f"Top 10 Countries: {decade}", fontsize=12)
    plt.xlabel("Country", fontsize=10)
    plt.ylabel("Population Increase (%)", fontsize=10)
    plt.xticks(rotation=45, ha='right')  # Rotate country names for better readability

# Adjust the layout and show the plot
plt.tight_layout()
plt.show()
```
